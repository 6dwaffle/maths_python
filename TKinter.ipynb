```python
# Program to make a simple

# login screen

import tkinter as tk
from tkinter import *
from PIL import ImageTk,Image
root=tk.Tk()
# setting the windows size
root.geometry("600x600")
root.title('Employee Details')
root.config(bg="black")



# declaring string variable
# for storing name and password
firstname_var=tk.StringVar()
lastname_var=tk.StringVar()
department_var=tk.StringVar()
email_var=tk.StringVar()
phoneno_int=tk.IntVar()
age_int=tk.IntVar()
ide_int=tk.IntVar()
salary_int=tk.IntVar()
# defining a function that will
# get the name and password and
# print them on the screen
def submit():
    
    firstname=firstname_var.get()
    lastname=lastname_var.get()
    department=department_var.get()
    email=email_var.get()
    phoneno=phoneno_int.get()
    age=age_int.get()
    salary=salary_int.get()
    ide =ide_int.get()
    
    print("Id of Employee is:" , ide)
    print("First name of Employee is : "+ firstname)
    print("Last name of Employee is : "+ lastname)
    print("The Department of Employee is : "+ department)
    print("The Email of Employee is : "+ email)
    print("The Age of Employee is : ", age)
    print("The Phone Number of the Employee is : ", phoneno)
    print("The Salary of the Employee is : ", salary)
    ide_int.set("")
    firstname_var.set("")
    lastname_var.set("")
    department_var.set("")
    email_var.set("")
    phoneno_int.set("")
    age_int.set("")
    salary_int.set("")

heading_label= tk.Label(root, text= 'Enter Employee Details', font=('Times New Roman',40, 'bold'), bg='red')
# creating a label for Id
ide_label= tk.Label(root, text= 'Id', font= ('Times New Roman',20,'bold'), bg='yellow')

# creating a entry for Id
ide_entry=tk.Entry(root, textvariable= ide_int, font= ('Times New Roman',20,'normal'), bg='yellow')    
    
# creating a label for
# name using widget Label
firstname_label= tk.Label(root, text= 'Firstname', font=('Times New Roman',20, 'bold'), bg='skyblue')

# creating a entry for input
# name using widget Entry
firstname_entry= tk.Entry(root,textvariable= firstname_var, font=('Times New Roman',20,'normal'), bg='skyblue')

# creating a label for
# last name using widget Label
lastname_label= tk.Label(root, text= 'Lastname', font=('Times New Roman',20, 'bold'), bg='pink')

# creating a entry for input
# last name using widget Entry
lastname_entry= tk.Entry(root,textvariable= lastname_var, font=('Times New Roman',20,'normal'), bg='pink')

# creating a label for Department
department_label= tk.Label(root, text= 'Department', font= ('Times New Roman',20,'bold'), bg='blue')

# creating a entry for Department
department_entry=tk.Entry(root, textvariable= department_var, font= ('Times New Roman',20,'normal'), bg='blue')

# creating a label for
# email using widget Label
email_label= tk.Label(root, text= 'Email', font=('Times New Roman',20, 'bold'), bg='green')

# creating a entry for input
# name using widget Entry
email_entry= tk.Entry(root,textvariable= email_var, font=('Times New Roman',20,'normal'), bg='green')

# creating a label for age
age_label= tk.Label(root, text= 'Age', font= ('Times New Roman',20,'bold'), bg='cyan')

# creating a entry for Age
age_entry=tk.Entry(root, textvariable= age_int, font= ('Times New Roman',20,'normal'), bg='cyan')

# creating a label for Phone Number
phoneno_label= tk.Label(root, text= 'Phone Number', font= ('Times New Roman',20,'bold'), bg='yellow')

# creating a entry for Phone Number
phoneno_entry=tk.Entry(root, textvariable= phoneno_int, font= ('Times New Roman',20,'normal'), bg='yellow')

# creating a label for Salary
salary_label= tk.Label(root, text= 'Salary', font= ('Times New Roman',20,'bold'), bg='skyblue')

# creating a entry for Salary
salary_entry=tk.Entry(root, textvariable= salary_int, font= ('Times New Roman',20,'normal'), bg='skyblue')

# creating a button using the widget
# Button that will call the submit function
sub_btn=tk.Button(root,text= 'Submit', command= submit)

# placing the label and entry in
# the required position using grid
# method
heading_label.grid(row=0,column=2)
ide_label.grid(row=1,column=1)
ide_entry.grid(row=1,column=2)
firstname_label.grid(row=2,column=1)
firstname_entry.grid(row=2,column=2)
lastname_label.grid(row=3,column=1)
lastname_entry.grid(row=3,column=2)
department_label.grid(row=4,column=1)
department_entry.grid(row=4,column=2)
email_label.grid(row=5,column=1)
email_entry.grid(row=5,column=2)
phoneno_label.grid(row=6, column=1)
phoneno_entry.grid(row=6, column=2)
age_label.grid(row=7, column=1)
age_entry.grid(row=7, column=2)
salary_label.grid(row=8, column=1)
salary_entry.grid(row=8, column=2)
sub_btn.grid(row=9,column=3)
# for the window to display
root.mainloop()
```
